```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
dataset = pd.read_excel('Data.xlsx.xlsx')
```

```python
dataset.head(5)
```

Regalos mas populares para cada grado de edad.

```python
Regalos = dataset.groupby(['gift_name'])['age'].mean().sort_values(ascending=False)
Regalos.head(5)
```

Encontrar los países donde los gatgets son mas solicitados

```python
filtro = dataset['gift_type'] == 'Electronics'
df_filtrado = dataset[filtro]
paises = df_filtrado.groupby('country')['country'].count().sort_values(ascending=False).head(3)
paises.head()
```

Regalo mas económico

```python
eco = dataset ['price'] < 20
econo = dataset[eco]
economico = econo.groupby(['gift_name'])['price'].count().sort_values(ascending=False)
economico.head(5)
```

```python
Regalos.plot(kind='bar', title='Regalos mas populares por edad', xlabel='Regalo', ylabel='Edad')
```

```python
paises.plot(kind='bar', title='Paises con mas gatgets solicitados', xlabel='Cantidad', ylabel='Paises')
```

```python
economico.plot(kind='bar', title='Regalo mas economico', xlabel='Regalo', ylabel='Precio')
```
